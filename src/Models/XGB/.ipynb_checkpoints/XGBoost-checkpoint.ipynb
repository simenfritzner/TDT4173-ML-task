```python

# Assuming y_a is a predefined DataFrame with the necessary columns
new_df = y_c.copy()  # Create a new DataFrame based on y_a
new_df["new_time"] = pd.to_datetime(new_df["time"], format='%Y-%m-%d %H:%M:%S')
new_df['hour'] = new_df['new_time'].dt.hour
new_df["month"] = new_df['new_time'].dt.month

# Filter for summer months
summer_months_df = new_df[new_df["month"].isin([5, 6, 7])]

# Count the total number of data points for each hour and month
total_count = summer_months_df.pivot_table(index='hour', columns='month', 
                                           values='pv_measurement', 
                                           aggfunc='count', 
                                           fill_value=0)

# Count the number of zeros in 'pv_measurements' for each hour and month
zeros_count = summer_months_df.pivot_table(index='hour', columns='month', 
                                           values='pv_measurement', 
                                           aggfunc=lambda x: (x == 0).sum(), 
                                           fill_value=0)

# Calculate the percentage of zeros
percentage_zeros = (zeros_count / total_count) * 100

print("Percentage of zeros in 'pv_measurements' for each hour during May, June, and July:")
print(percentage_zeros)
```
